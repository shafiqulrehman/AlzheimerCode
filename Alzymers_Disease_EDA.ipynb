```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the CSV file
file_path = "/content/data.csv"  # Replace with the actual file path
df = pd.read_csv(file_path)

# Step 2: Remove the first and last column
df = df.iloc[:, 1:-1]

# Step 3: Create box plots before outlier detection (for the first 18 columns)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Box Plot Before Outlier Detection")
df.iloc[:, :18].boxplot()
plt.xticks(rotation=90)

# Step 4: Apply the IQR method to detect outliers
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Detect outliers
outliers = ((df < lower_bound) | (df > upper_bound))

# Step 5: Replace outliers with the mode of their respective columns
for col in df.columns:
    df[col] = df[col].mask(outliers[col], df[col].mode()[0])

# Step 6: Create box plots after outlier replacement (for the first 18 columns)
plt.subplot(1, 2, 2)
plt.title("Box Plot After Outlier Replacement")
df.iloc[:, :18].boxplot()
plt.xticks(rotation=90)

# Show the box plots
plt.tight_layout()
plt.show()

# Step 7: Save the cleaned dataset to a new CSV file
cleaned_file_path = "cleaned_data.csv"  # Replace with the desired file path
df.to_csv(cleaned_file_path, index=False)
print(f"Cleaned dataset saved to {cleaned_file_path}")
```
